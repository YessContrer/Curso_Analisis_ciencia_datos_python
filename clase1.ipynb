## Python sirve como una calculadora básica

- $+$ --> suma
- $-$ --> resta
- $*$ --> multiplicación
- $**$ --> potencia
- $/$ --> división
- $//$ --> división al piso
- '%' --> residuo de división



```python
t = (2+5, 2-5, 2*5, 2**5, 2/5, 2//5, 2%5)  #Se genera una una tupla
t
print(t) #se imprime la tupla "t"
```

#### La función range()

`Es un generador de índices.`




```python
range(1,11) #se hace un indice
```

Para vsualizar los elementos de un generador se encesitan ciclos

"for"



```python
for i in range(3,10): #ciclo for para mostrar los valores de i dentro del rango de 3 a 9 (10-1=9)
  print(i)
```

```python
L3 = [ i for i in range(11)] #lista (L3) del ciclo for para mostrar los valores de i en el rango del 0 al 10 ( 11-1 = 10 )
L3
```
