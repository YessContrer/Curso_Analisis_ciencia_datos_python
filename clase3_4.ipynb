```python
#cargar el modulo/biblioteca "numpy" para usar funciones...
#que python no tiene precargadas

import numpy as np

#establece una semilla para el generador de numeros aleatorios...
#hace que siempre salgan los mismos numeros aleatorios

np.random.seed(1)

#se genera un diccionario con 4 keys y dentro de la funcion (10 numeros de
#ensayos, con 0.5 de probabilidad, 5 repeticiones), se tendran 5
#valores aleatorios en cada entrada del diccionario

d = {'A': np.random.binomial(10,0.5,5),
     'B': np.random.binomial(10,0.3,5),
     'C': np.random.binomial(10,0.6,5),
     'D': np.random.binomial(10,0.4,5)}

#type(d)

#d.values() #valores de d
#d.keys()   #keys de d
#d.items()  #items de d

d #impresion de "d"
```

```python
#carga el modulo/biblioteca "pandas" para usar funciones que python no...
#tiene precargadas

import pandas as pd

#creacion de una lista para usar como indices en un dataframe
idx = [1,3,5,7,9]

df = pd.DataFrame(d) #deja el dataframe como el que es
df = pd.DataFrame(d,index = idx) #cambia el nombre de indices de la...
#dataframe puede ser strings

df
```

```python
df["Promedio"] = df.mean(axis=1) #promedio de los renglones (axis=1) del...
#dataframe
df
```

```python
wd_1 = wd2[['class','Alcohol','Total phenols','Flavanoids','Color intensity']]
#del data frame se selecciona las columnas que se necesitan y se colocan
#en wd_1
wd_1
```

```python
alcohol = wd_1['Alcohol'] #seleccion de los datos en la columna alcohol
alcohol
alcohol_ = wd_1[wd_1['Alcohol']> 13.2] #seleccion de los datos de toda...
#la dataframe que satisfacen la condicion de que el valor en la columna...
#"Alcohol" sea mayor a 3.2
alcohol_

```

```python
#cambio de nombre de 1 columna
alcohol_ = alcohol_.rename(columns={"Color intensity":"Color"})

#cambio de nombre de 1 o mas columnas
alcohol_ = alcohol_.rename(columns={ "Total phenols":"Phenols","Alcohol":"Alc"})
alcohol_
```

```python
alcohol_2 = alcohol_[alcohol_["Color"] > 0 ] #valores mayores a cero
alcohol_2

```

```python
alcohol_2.info() #informacion del dataframe
```

```python
#agrupacion de datos por categoria
g_alcohol_2 = alcohol_2.groupby(by = "class")['Color'].count()
g_alcohol_2

#ordenar los valores en orden descendiente (valores)
g2 = g_alcohol_2.sort_values(ascending = False)
g2

#ordenar los valores en orden descendiente (indices)
g3 = g_alcohol_2.sort_index(ascending = False)
g3
```

```python
#visualizacion de datos
#carga el modulo o biblioteca "seaborn" y "matplotlib.pyplot" para usar...
#funciones que python no tiene precargadas para graficar
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#crea un diccionario
datos = {"Sexo": ["F","F","M","M"],
         "Edad": ["18-21", "22-25", "18-21", "22-25"],
         "Value": [6.5, 6, 8.2, 6.1]}

df_2 = pd.DataFrame(datos) #crea el dataframe

plt.figure(figsize = (8,6)) #crea una figura de 8 por 6 pulgadas
sns.barplot(df_2,x = "Edad", y = "Value")  #crea una grafico de barras...
#con x datos de edad y y los valores
plt.ylabel("Valor promedio") #crea los labels de y
plt.xlabel("Edad") #crea los labels de x

plt.tight_layout() #ajusta automaticamente el diseño del grafico para que no...
#se encimen ni corten los elementos como titulos, ejes, etiquetas, leyendas...
plt.show() #muestra la figura en pantallaa
```
